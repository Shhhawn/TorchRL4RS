```python
import pandas as pd
import os
```

```python
item_info_path = "../data/rl4rs_benchmark_materials/raw_data/item_info.csv"

if os.path.exists(item_info_path):
    print("="*20, "Item Info Probe", "="*20)
    with open(item_info_path, 'r') as f:
        header = f.readline().strip()
        print(f"表头：{header}")
        for i in range(2):
            print(f"样例 {i+1}: {f.readline().strip()[:100]}...") # 只截取前100字符，防止emb太长刷屏
else:
    print(f"未找到 {item_info_path}")
```

```python
rl_file = "../data/rl4rs_benchmark_materials/raw_data/small_rl4rs_dataset_a_rl.csv"

if os.path.exists(rl_file):
    print("\n" + "="*20, "2. RL Log 探测", "="*20)
    with open(rl_file, 'r', encoding='utf-8') as f:
        header = f.readline().strip()
        print(f"表头: {header}")

        first_row = f.readline().strip()
        print("\n--- 第一条数据拆解 ---")
        columns = header.split('@')
        # 工业界数据如果内部有逗号，通常用特殊分隔符（如 \t, @, ;），我们先暴力按逗号切分看看是否错位
        values = first_row.split('@')

        print(f"表头期望列数: {len(columns)}")
        print(f"实际切分列数: {len(values)}")
        
        # 打印前 5 个字段的对应关系
        print("\n前几个字段的值:")
        for col, val in zip(columns, values):
            print(f"[{col}]: {val}")

else:
    print(f"未找到 {rl_file}")
```
